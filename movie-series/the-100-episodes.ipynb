# The 100 Episodes

Library dan/atau module yang digunakan dalam latihan kali ini ialah:

- request (seperti httr pada R)
- BeautifulSoup (seperti rvest pada R)
- pandas (seperti tibble pada R)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_The_100_episodes")
```

```python
page = BeautifulSoup(res.text, "html.parser")
```

```python
episode_tables = page.find_all("table", class_ = "wikitable plainrowheaders wikiepisodetable")
```

```python
episodes = []
for table in episode_tables:
    headers = []
    for header in table.find("tr").find_all("th"):
        headers.append(header.text)
    for row in table.find_all("tr", class_ = "vevent"):
        values = []
        value = row.find_all("td")
        for col in row.find_all(["th", "td"]):
            values.append(col.text)
        episodes_dict = { headers[i] : values[i] for i in range(len(values))}
        episodes.append(episodes_dict)
```

```python
df = pd.DataFrame(episodes)
df
```

```python
# save to csv
df.to_csv("./the-100-episodes.csv", index = False)
```
